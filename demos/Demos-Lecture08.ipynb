```python
import numpy as np
import matplotlib.pyplot as plt
```

## Demo 1: Shifted square waves

Our original square wave, expressed as a Fourier series of sine terms only.

```python
def coeffs_sin(k):
    return 4 / (k * np.pi) if k % 2 == 1 else 0

def square_sin(t, max_k):
    return np.sum([coeffs_sin(k) * np.sin(k * t) for k in range(0, max_k+1)])
```

```python
plt.figure(figsize=(12, 8))
t_range = np.linspace(-3*np.pi, 3*np.pi, 1000)
plt.plot(t_range, [np.real(square_sin(t, 1000)) for t in t_range])
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
```

The same series, but we expressed the coefficients as complex exponentials.

```python
def coeffs_exp(k):
    return 2 / (1j * k * np.pi) if k % 2 == 1 else 0

def square_exp(t, max_k):
    return np.sum([coeffs_exp(k) * np.exp(1j * k * t) for k in range(-max_k, max_k+1)])
```

```python
plt.figure(figsize=(12, 8))
t_range = np.linspace(-3*np.pi, 3*np.pi, 1000)
plt.plot(t_range, [np.real(square_exp(t, 1000)) for t in t_range])
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
```

Shift the signal left by $\pi/2$ by adding a complex phase to the Fourier coefficients.

```python
def coeffs_shifted(k):
    return 2 * np.sin(k * np.pi / 2) / (k * np.pi) if k % 2 == 1 else 0

def square_shifted(t, max_k):
    return np.sum([coeffs_shifted(k) * np.exp(1j * k * t) for k in range(-max_k, max_k+1)])
```

```python
plt.figure(figsize=(12, 8))
t_range = np.linspace(-3*np.pi, 3*np.pi, 1000)
plt.plot(t_range, [np.real(square_shifted(t, 1000)) for t in t_range])
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
```

Perform the time scaling by changing the frequency of the complex exponential signals.

```python
def coeffs(k):
    return 2 * np.sin(k * np.pi / 2) / (k * np.pi) if k % 2 == 1 else 0

def square_shifted_scaled(t, max_k):
    return np.sum([coeffs(k) * np.exp(2j * k * t) for k in range(-max_k, max_k+1)])
```

```python
plt.figure(figsize=(12, 8))
t_range = np.linspace(-3*np.pi, 3*np.pi, 1000)
plt.plot(t_range, [np.real(square_shifted_scaled(t, 1000)) for t in t_range])
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
```

## Demo 2: DT Fourier series coefficients

Example of a non-periodic DT signal. 

```python
n_range = np.arange(-60, 60)
plt.figure(figsize=(12, 8))
plt.stem(n_range, np.cos(3 * n_range))
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
```

This one is periodic; the argument of $\cos(5\pi n/6)$ will become a multiple of $2\pi$ at $N=12$.

```python
n_range = np.arange(-20, 20)
plt.figure(figsize=(12, 8))
plt.stem(n_range, np.cos(5 * np.pi * n_range / 6))
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
```

Another periodic example $( N =3)$

```python
n_range = np.arange(-20, 20)
plt.figure(figsize=(12, 8))
plt.stem(n_range, np.sin(2 * np.pi * n_range / 3))
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
```

The sum of the two signals is also periodic, with period $N = 12$ (since 3 is a factor of 12, 4 copies of the $\sin$ term "fit" inside the $\cos$ term.

```python
n_range = np.arange(-20, 20)
plt.figure(figsize=(12, 8))
plt.stem(n_range, np.sin(2 * np.pi * n_range / 3) + np.cos(5 * np.pi * n_range / 6))
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
```
